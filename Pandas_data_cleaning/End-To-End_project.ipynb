```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df= df.rename(columns = {'Purchase_Amount_(USD)':'Purchase_Amount'})
```

#creating a new column

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
```

```python
df['frequency_of_purchases_days'] = df['frequency_of_purchases'].map(frequency_mapping) 
```

```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Define your server and database
# Note: the 'r' before the string tells Python to treat the backslash literally
server = r"localhost\SQLEXPRESS01" 
database = "project"

# 2. Create the MS SQL Server connection string using Windows Authentication
# (Trusted_Connection=yes handles your Windows login automatically)
connection_url = f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&Trusted_Connection=yes"

# 3. Create the engine
engine = create_engine(connection_url)


# 4. Write DataFrame to SQL Server
table_name = "customer"
df.to_sql(table_name, engine, if_exists="replace", index=False)

# 5. Read back sample
test_df = pd.read_sql("SELECT TOP 5 * FROM customer", engine) # Note: MS SQL uses 'TOP 5' instead of 'LIMIT 5'
test_df
```
